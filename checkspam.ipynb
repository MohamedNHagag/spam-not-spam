```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df=pd.read_csv(r'D:\data science\7)machine learning\projects\classfication logistic & SVM\5) spam & not spam\mail_data.csv')
```

```python
df.head()
```

```python
df['Category']=df['Category'].replace({'spam':0,'ham':1})
```

```python
df['Category'].value_counts().sort_values(ascending=True)
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.info()
```

```python
x=df['Message']
y=df['Category']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=3,stratify=y,test_size=0.1)
```

```python
#convert text in category to numercal data

vectorizer=TfidfVectorizer()
x_train_vectorizer=vectorizer.fit_transform(x_train)
x_test_vectorizer=vectorizer.transform(x_test)

#certainaly output became int
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
#create model and training
model=LogisticRegression()
model.fit(x_train_vectorizer,y_train)
```

```python
#prediction x_train and accuracy score
predict_train=model.predict(x_train_vectorizer)
accuracy_train=accuracy_score(y_train,predict_train)
accuracy_train
```

```python
#prediction x_test and accuracy score
predict_test=model.predict(x_test_vectorizer)
accuracy_test=accuracy_score(y_test,predict_test)
accuracy_test
```
